```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('adult.csv')
df.columns=['age','workclass','fnlwgt','education','education-num','marital-status','occupation','relationship','race','sex','capital-gain','capital-loss','hours-per-week','native-country','Target']
df.head()
```

```python
df.isna().sum().sum()
```

```python
# Histogram of age
plt.figure(figsize=(8, 6))
plt.hist(df['age'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Histogram of Age')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(x='education', bins=20, kde=True,data=df)
plt.xlabel('Education')
plt.ylabel('Frequency')
plt.title('Histogram of Education Distribution')
plt.show()
```

```python
# Dot plot of age by income
plt.figure(figsize=(8, 6))
sns.stripplot(x='Target', y='age', data=df.head(100),hue='Target')
plt.xlabel('Income')
plt.ylabel('Age')
plt.title('Dot Plot of Age by Income')
plt.show()
```

```python
#Bar Plot
plt.figure(figsize=(8, 6))
df['education'].value_counts().plot(kind='bar', color='pink')
plt.xlabel('Education level')
plt.ylabel('Count')
plt.title('Education Level Distribution')
plt.show()
```

```python
# Use a bar plot to visualize income levels by age
plt.figure(figsize=(10, 6))
sns.barplot(x="age",y="education",data=df)
plt.xlabel('Age')
plt.ylabel('Distribution of Income Levels by Age')
plt.title('Distribution of Income Levels by Age')
plt.show()
```

```python
# Line Plot

plt.figure(figsize=(10, 6))
plt.plot(df['age'].head(100), df['Target'].head(100), marker='o')
plt.xlabel('Age')
plt.ylabel('Income')
plt.title('Line Chart of Age by Income')
plt.grid(True)
plt.show()
```

```python
# Line Plot
plt.figure(figsize=(10, 6))
sns.lineplot(x="age", y="education", data=df)
plt.xlabel('Age')
plt.ylabel('Education')
plt.title('Education by Age')
plt.show()
```

```python
plt.figure(figsize=(8, 15))

# Box Plot of Age by Income
plt.subplot(3, 1, 1)
sns.boxplot(x='Target', y='age', data=df)
plt.title('Box Plot of Age by Income')

# Histogram of Hours per Week
plt.subplot(3, 1, 2)
sns.histplot(x='hours-per-week', data=df, kde=True)
plt.title('Histogram of Hours per Week')

# Scatter Plot of Education Num by Hours per Week
plt.subplot(3, 1, 3)
sns.scatterplot(x='hours-per-week', y='education-num', data=df, hue='Target')
plt.title('Scatter Plot of Education Num by Hours per Week')

plt.tight_layout()
plt.show()
```

```python
#Pie Chart
plt.figure(figsize=(8, 6))
df['education'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Distribution of Target Variable')
plt.show()
```

```python
# Scatterplot:
sns.scatterplot(x="hours-per-week", y="age", data=df,hue='hours-per-week')
```

```python
# Adding Box Plots to a Scatter Plot
plt.figure(figsize=(8, 6))
sns.scatterplot(x='age', y='education-num', data=df, hue='Target')
sns.boxplot(x='age', y='education-num', data=df, color='red', linewidth=2)
plt.title('Scatter Plot with Box Plots')
plt.show()
```
